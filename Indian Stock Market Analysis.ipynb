# Working on Data

```python
#Converting the “Date” column dtype from object to date
#com["Date"]=pd.to_datetime(com["Date"])
com["Date"] = pd.to_datetime(com["Date"], format='%d-%m-%Y')  # Adjust the format as per your actual date format
```

```python
com.info()
```

```python
#Dropping columns
com=com.drop(['col_name1','col_name2','col_name3'], axis=1)
```

```python
#Adding new columns to the dataset
com['Day']=com["Date"].dt.day
com['Month']=com["Date"].dt.month
com['Year']=com["Date"].dt.year
```

```python
com.head()
```

# Comparing the Data

```python
#Price Comparision
symbols = com['Symbol'].unique()  # Get unique symbols in the 'Symbol' column

plt.figure(figsize=(20, 6))

# Define colors for each symbol
symbol_colors = {
    'TCS': 'blue',
    'Wipro': 'green',
    'Infosys': 'orange'
}

# Plot data for each symbol
for symbol in symbols:
    symbol_data = com[com['Symbol'] == symbol]
    color = symbol_colors.get(symbol)
    plt.plot(symbol_data['Date'], symbol_data['Open'], label=symbol, color=color)

plt.title("Price Comparison between Indian Stock Companies")
plt.xlabel("Year")
plt.ylabel("Price")
plt.legend(title="Company Names")
plt.show()
```

### According to the graph above, the price of TCS has skyrocketed significantly higher than that of Wipro and Infosys. TCS’s pricing trajectory has been generally upward from its beginning, whereas Wipro and Infosys have been more on a consolidation trend.

# Return on Investment (ROI)

```python
import matplotlib.pyplot as plt

# Assuming you have loaded your data into a DataFrame named 'data'
# You can modify the code as follows:

symbols = com['Symbol'].unique()  # Get unique symbols in the 'Symbol' column

# Define colors for each symbol
symbol_colors = {
    'TCS': 'blue',
    'Wipro': 'green',
    'Infosys': 'orange'
}
# Initialize lists to store ROI values and corresponding symbols
rois = []
symbols_list = []
# Calculate ROI for each symbol
for symbol in symbols:
    symbol_data = com[com['Symbol'] == symbol]
    
    # Calculate ROI based on the opening prices
    initial_investment = symbol_data['Open'].iloc[0]  # Initial investment is the first day's opening price
    current_value = symbol_data['Open'].iloc[-1]  # Last day's opening price
    roi = ((current_value - initial_investment) / initial_investment) * 100
    
    # Append ROI and symbol to lists
    rois.append(roi)
    symbols_list.append(symbol)

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(rois, labels=symbols_list, colors=[symbol_colors.get(symbol, 'black') for symbol in symbols_list], autopct='%1.1f%%', startangle=140)

plt.title("ROI Distribution between Symbols")
plt.show()
```
